```python
# Data_formats

## 21/04/2026

#  a simple dict
data = {
    "name": "sargam",
    "age" : 20,
    "city" : "shimla"
}

print(data)
print(type(data))
```

```python
# converting to json
import json 
json_data = json.dumps(data)
print(json_data)
print(type(json_data))
```

```python
## memory taken by json file
import sys

size_json = sys.getsizeof(json_data)
print(size_json)   ## it is in bytes
```

```python
# binary conversion
binary_data = bytes(json_data,'utf-8')
print(binary_data)
print(type(binary_data))
```

```python
## memory size of binary file
size_binary = sys.getsizeof(binary_data)
size_binary            ## binary data takes less space compare to json
```

```python
# Pandas
import pandas as pd

# Creating a basic DataFrame
data = {
    'image_id': ['img_001', 'img_002', 'img_003', 'img_004'],
    'species': ['Panthera pardus fusca', 'Panthera pardus fusca', 'Unknown', 'Panthera pardus fusca'],
    'verified': [True, True, False, True],
    'confidence_score': [0.98, 0.88, 0.45, 0.92]
}

df = pd.DataFrame(data)

# Display the DataFrame
print(df)
```

```python
# view the first 2 rows of the dataset
print(df.head(2))

# get a quick summary of the columns, non-null counts, & data types
print(df.info())

# get basic math statistics (mean, min, max) for numerical columns
print(df.describe())
```

```python
# Data_formats

## 21/04/2026

#  a simple dict
data = {
    "name": "sargam",
    "age" : 20,
    "city" : "shimla"
}

print(data)
print(type(data))
```

```python
# converting to json
import json 
json_data = json.dumps(data)
print(json_data)
print(type(json_data))
```

```python
## memory taken by json file
import sys

size_json = sys.getsizeof(json_data)
print(size_json)   ## it is in bytes
```

```python
# binary conversion
binary_data = bytes(json_data,'utf-8')
print(binary_data)
print(type(binary_data))
```

```python
## memory size of binary file
size_binary = sys.getsizeof(binary_data)
size_binary            ## binary data takes less space compare to json
```
